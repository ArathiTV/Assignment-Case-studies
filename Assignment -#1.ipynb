### 5.    Perform meaningful visualization using the dataset.Bring at least 3  visualizations

### 1.  bar plot for 3 classification

```python
# set the figuresize
plt.figure(figsize=(8,7))
# creating bar plot
plt.bar(data['Classification'],data['SL'])
# set the title,y-axis ,x-axis with fontsize
plt.title('Bar plot for 3 Classification',fontsize=19)
plt.ylabel('Sepal Length',fontsize=15)
plt.xlabel('Classification', fontsize=15)
# show the plot
plt.show()
```

### 2.  Scatter plot of PL and PW

```python
# creating the scatterplot
plt.scatter(data['PL'],data['PW'])
# set title,y-axis,x-axis with fontsize
plt.title('scatter plot of PL and PW',fontsize=20)   
plt.ylabel('y-axis',fontsize=15)
plt.xlabel('x-axis',fontsize=15)
# show the plot
plt.show()
```

### 3.  Bar plot of petal_width and petal_length

```python
# set the figure size
plt.figure(figsize=(7,6))
# creating the bar plot
plt.bar(data['PW'], data['PL'],color=['yellow'])
# set the title,x-axis,y-axis with fontsize
plt.xlabel('Petal width',fontsize=15)
plt.title('petal width and petal length',fontsize=19)
plt.ylabel('Petal length',fontsize= 15)
# show the plot
plt.show()
```

### 4. Seaborn Plots

```python
# creating the pairsplot in seaborn
sns.pairplot(data)
# show the plot
plt.show()
```

### 5. box plot 

```python
# set figure size
plt.figure(figsize=(10,7))
# creating the boxplot , #  5 summury statistic
plt.boxplot([data['SL'],data['SW'],data['PL'],data['PW']])
# set the title with fontsize
plt.title('Box plots Iris Features',fontsize=22)
# show the plot
plt.show()
```

### 6. Line plot of Iris dataset

```python
# set the figure size
plt.figure(figsize=(29,25))
# creating line plot
data.plot()
# set title and legend
plt.title('Iris Dataset',fontsize=19)
plt.legend()
# show the plot
plt.show()
```
